# **Stop Words in English**

```python
import nltk
nltk.download("all")
from nltk.corpus import stopwords
```

```python
#Reading the stops words file
en_stops = set(stopwords.words('english'))
print(en_stops)
```

```python
Sentence1= """ Once there was a Lion in the jungle who used to kill 2-3 animals daily for his meal. 
All animals went to him to tell, that daily one of them will come to him for his meal.
So, the Lion agreed and this started going for many days. One day, it was Rabbit’s turn. When he was on his way he saw a well.
Now he plans to kill the lion and save himself. He went to the lion and told him that, there is another lion who claims to be more powerful than him.
Then the lion asks the rabbit to take him to that lion. The rabbit takes him to the well and said he lives here. 
When the lion looked in the well he saw his own reflection and jumped in the well and dies."""
print("varsha")
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
#Reading the English stop words file
stop_words=set(stopwords.words("english"))
 
#Converting the data into word tokenizer so each word will treated individually
word_tokens = word_tokenize(Sentence1)

#Creating the list to store the value of stop word in sentence
stop_word_remove = []

#Converting the word_token into lower case because stope word file have all the word in lower case 
word_tokens_lower=[]
for w in word_tokens:
  w= w.lower()
  word_tokens_lower.append(w)

#Creating the list to store the value of stop words in sentence
stop_word_remove=[]


#Creating the list to store the value of sentence without stop words
sentence_wo_stop_word=[]
 
for w in word_tokens_lower:
    if w in stop_words:
        stop_word_remove.append(w)
    else :
        sentence_wo_stop_word.append(w)
print("varsha")
print("Original Sentence", word_tokens_lower)
print("Stop word in the Sentence", stop_word_remove)
print("Original Sentence without stop word",sentence_wo_stop_word)
```
